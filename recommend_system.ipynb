# Demographic Filtering
Đưa ra các đề xuất tổng quát cho mọi người dùng, dựa trên mức độ phổ biến, hoặc điểm đánh giá cao của hầu hết mọi người. Vì mỗi người dùng là khác nhau nên cách tiếp cận này được coi là quá đơn giản. Tuy nhiên, nó cũng đem lại hiệu quả cho những người mới về những bộ phim được rating cao trên bản xếp hạng, đem lại hứng thú và tò mò xem phim. Ý tưởng cơ bản đằng sau hệ thống này là những bộ phim nổi tiếng hơn và được giới phê bình đánh giá cao hơn sẽ có xác suất được khán giả trung bình yêu thích cao hơn.

IMDB's weighted rating được sử dụng để đánh giá chất lượng của các bộ phim. Hệ thống này được sử dụng bởi trang web Internet Movie Database (IMDB) để giúp người xem có cái nhìn tổng quan về sự phổ biến và đánh giá của một tác phẩm cụ thể.

- v: vote_count của một bộ phim
- m: số lượng vote_count tối thiểu để đánh giá một bộ phim (ngưỡng tùy chỉnh)
- R: vote_average của một bộ phim
- C: trung bình vote_average của tất cả bộ phim

```python
C = df2['vote_average'].mean()
C
```

```python
df2['vote_count'].describe()
```

```python
# để một bộ phim có mặt trong bảng xếp hạng thì số phiếu bầu cao hơn từ 90% phim
m = df2['vote_count'].quantile(0.9)
m
```

```python
q_movie = df2[df2['vote_count'] >= m]
q_movie.shape
```

```python
weight = []
for v, R in zip(q_movie['vote_count'],q_movie['vote_average']):
    weight.append((R*v + C*m) / (v+m))
q_movie['weight_rating'] = weight
```

```python
q_movie = q_movie[['id', 'title', 'vote_count', 'vote_average', 'weight_rating']]
q_movie = q_movie.sort_values('weight_rating', ascending=False)
```

```python
import json
def get_info(x, num_feature=3):
    lst = []
    if x:
        for i in json.loads(x):
            lst.append(i['name'])
            if len(lst) == num_feature:
                break
    return ' '.join(lst)
```

```python
overviews = []
for i in range(len(df2['overview'])):
    overviews.append(' '.join([df2['title'][i], df2['overview'][i], get_info(df2['genres'][i]),
                               get_info(df2['cast'][i]), get_info(df2['keywords'][i]), df2['tagline'][i]]))
```

```python
from nltk.corpus import stopwords
from nltk.stem.lancaster import LancasterStemmer
from pyvi import ViTokenizer

def Preprocessing(overview):
    overview = overview.lower()
    punctuation = '''!"#$%&\'()*+,.-/:;<=>?@[\\]^_`{|}~'''
    for punc in punctuation:
        if punc == "'":
            overview = overview.replace(punc, '')
        else: overview = overview.replace(punc, ' ')
    overview = ' '.join([word for word in overview.split() if word not in stopwords.words('english')])
    overview = ' '.join([LancasterStemmer().stem(word) for word in overview.split() if len(word) > 1])
    return overview
```

```python
overviews = [Preprocessing(overview) for overview in overviews]
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def cosine_similar(x, lst):
    cosine = cosine_similarity([x],lst)
    return np.argsort(cosine)[0][::-1]
```

# Collaborative Filtering
Ý tưởng cơ bản của thuật toán này là dự đoán mức độ yêu thích của một user đối với một item dựa trên các users khác “gần giống” với user đang xét. Việc xác định độ “giống nhau” giữa các users có thể dựa vào mức độ quan tâm (rating) của các users này với các items khác mà hệ thống đã biết trong quá khứ

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import mean_squared_error
from tqdm import tqdm

ratings = pd.read_csv('/content/drive/MyDrive/CV/Recommend System/The Movies Dataset/ratings_small.csv')
```

```python
ratings
```

```python
num_user = len(ratings['userId'].unique())
num_movie = len(ratings['movieId'].unique())
print("Số lượng người dùng:", num_user)
print("Số lượng bộ phim:", num_movie)
```

```python
# format: index_rating = [[user_1, list_movie, list_rating], ..., [user_n, list_movie, list_rating]]
index_rating = [[] for _ in range(num_user)]
index_rating_test = [[] for _ in range(num_user)]
mean_rating = []
```

```python
def train_test_split(lst, train_size=0.8):
  lst_train = lst[:int(len(lst)*train_size)]
  lst_test = lst[int(len(lst)*train_size):]
  return lst_train, lst_test
```

```python
movieId = list(np.sort(ratings['movieId'].unique()))
index_movieId = []

for i in range(num_user):
    train, test = train_test_split(ratings[ratings['userId']==i+1])
    index_rating[i].append(i+1)
    index_rating[i].append(list(train['movieId']))
    mean_rating.append(round(np.average(train['rating']),2))
    index_rating[i].append([round(rate - mean_rating[-1],2) for rate in train['rating']])

    index_rating_test[i].append(i+1)
    index_rating_test[i].append(list(test['movieId']))
    index_rating_test[i].append(list(test['rating']))

    movie_id = list(np.sort(index_rating[i][1]))
    idx_movie = [movieId.index(x) for x in movie_id]
    index_movieId.append([movie_id, idx_movie])
```

```python
def convert_vector(idx_movieId, lst_rating):
    vec=[0]*len(movieId)
    for i in range(len(idx_movieId)):
        vec[idx_movieId[i]] = lst_rating[i]
    return vec
```

```python
vector_users = []
for i in range(len(index_rating)):
    vector_users.append(convert_vector(index_movieId[i][1], index_rating[i][2]))
```

```python
from sklearn.decomposition import TruncatedSVD
svd = TruncatedSVD(n_components=20)
vector_users_svd = svd.fit_transform(vector_users)
```

```python
movie_watched_user = []
for i in range(len(movieId)):
    movie_watched_user.append(list(ratings[ratings['movieId']==movieId[i]]['userId']))
```

```python
print(f'Movie_Id: {movieId[3]}, được xem và đánh giá bới các User_ID: {movie_watched_user[3]}')
```

```python
def get_cosine_similar(vector, lst_vector, k=3):
    cosine = cosine_similarity([vector],lst_vector)
    sim_sorted = np.sort(cosine[0])[::-1]
    idx_sorted = np.argsort(cosine[0])[::-1]
    return idx_sorted[:k], sim_sorted[:k]
```

```python
import time
```

```python
y_test, y_pred, start = [], [], time.time()
for i in tqdm(range(len(index_rating))):
    temp1, temp2, vec_current = [], [], vector_users_svd[i]
    for j in range(len(movieId)):
        if int(movieId[j]) in index_rating_test[i][1] and int(movieId[j]) not in index_rating[i][1]:
            lst_vec = []
            for idx in movie_watched_user[j]:
                lst_vec.append(vector_users_svd[idx-1])
            idx_vec, sim_vec = get_cosine_similar(vec_current, lst_vec)
            rating_of_users = [vector_users[movie_watched_user[j][idx]-1][j] for idx in idx_vec]
            result = np.dot(sim_vec, rating_of_users) / (np.sum(np.abs(sim_vec))+1e-20) + mean_rating[i]
            temp1.append(movieId[j])
            temp2.append(round(result,2))
    for v in range(len(index_rating_test[i][1])):
      y_test.append(index_rating_test[i][2][v])
      y_pred.append(temp2[temp1.index(index_rating_test[i][1][v])])

rms = mean_squared_error(y_test, y_pred, squared=False)
print(f'\nRMSE = {rms}, TIME: {time.time()- start}')
```

```python
recommend_movieId, result_rating = [], []
for i in range(len(index_rating)):
    temp1, temp2, vec_current, start = [], [], vector_users_svd[i], time.time()
    for j in range(len(movieId)):
        if int(movieId[j]) not in index_rating[i][1]:
            lst_vec = []
            for idx in movie_watched_user[j]:
                lst_vec.append(vector_users_svd[idx-1])
            idx_vec, sim_vec = get_cosine_similar(vec_current, lst_vec)
            rating_of_users = [vector_users[movie_watched_user[j][idx]-1][j] for idx in idx_vec]
            result = np.dot(sim_vec, rating_of_users) / (np.sum(np.abs(sim_vec))+1e-20) + mean_rating[i]
            temp1.append(movieId[j])
            temp2.append(round(result,2))
    recommend_movieId.append(temp1)
    result_rating.append(temp2)
    print(f'RECOMMEND USER: {index_rating[i][0]}, MOVIE_ID: {[temp1[idx] for idx in np.argsort(temp2)[::-1][:5]]}, TIME: {time.time()-start}')
```
