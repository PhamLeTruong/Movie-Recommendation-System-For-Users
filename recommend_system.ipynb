# Demographic Filtering
Đưa ra các đề xuất tổng quát cho mọi người dùng, dựa trên mức độ phổ biến, hoặc điểm đánh giá cao của hầu hết mọi người. Vì mỗi người dùng là khác nhau nên cách tiếp cận này được coi là quá đơn giản. Tuy nhiên, nó cũng đem lại hiệu quả cho những người mới về những bộ phim được rating cao trên bản xếp hạng, đem lại hứng thú và tò mò xem phim. Ý tưởng cơ bản đằng sau hệ thống này là những bộ phim nổi tiếng hơn và được giới phê bình đánh giá cao hơn sẽ có xác suất được khán giả trung bình yêu thích cao hơn.

IMDB's weighted rating được sử dụng để đánh giá chất lượng của các bộ phim. Hệ thống này được sử dụng bởi trang web Internet Movie Database (IMDB) để giúp người xem có cái nhìn tổng quan về sự phổ biến và đánh giá của một tác phẩm cụ thể.

- v: vote_count của một bộ phim
- m: số lượng vote_count tối thiểu để đánh giá một bộ phim (ngưỡng tùy chỉnh)
- R: vote_average của một bộ phim
- C: trung bình vote_average của tất cả bộ phim

```python
C = df2['vote_average'].mean()
C
```

```python
df2['vote_count'].describe()
```

```python
# để một bộ phim có mặt trong bảng xếp hạng thì số phiếu bầu cao hơn từ 90% phim
m = df2['vote_count'].quantile(0.9) 
m
```

```python
q_movie = df2[df2['vote_count'] >= m]
q_movie.shape
```

```python
weight = []
for v, R in zip(q_movie['vote_count'],q_movie['vote_average']):
    weight.append((R*v + C*m) / (v+m))
q_movie['weight_rating'] = weight
```

```python
q_movie = q_movie[['id', 'title', 'vote_count', 'vote_average', 'weight_rating']]
q_movie = q_movie.sort_values('weight_rating', ascending=False)
```

# Content Based Filtering
Đề xuất các mặt hàng tương tự dựa trên một mặt hàng cụ thể. Dựa trên nội dung phim(overview, cast, crew, keyword, tagline, ...) để tìm điểm tương đồng của nó với các phim khác. Ý tưởng chung đằng sau các hệ thống giới thiệu này là nếu một người thích một mặt hàng cụ thể, người đó cũng sẽ thích một mặt hàng tương tự như vậy.

```python
df2['overview'].head()
```

```python
# phát hiện có 3 ô không có overview của phim
print('nan in overview: ',df2['overview'].isnull().sum())
# thay thế các giá trị nan thành chuỗi rỗng
df2['overview'] = df2['overview'].fillna('')
```

Đối với các phim không có overview nói riêng và tổng thể các phim nói chung, ta thêm thông tin bổ sung vào overview như genres, title, cast, keyword, tagline để bổ sung mô tả cho phim 

```python
# kiểm tra thông tin nan trong các trường
print('nan in genres: ',df2['genres'].isnull().sum())
print('nan in title: ',df2['title'].isnull().sum())
print('nan in cast: ',df2['cast'].isnull().sum())
print('nan in keywords: ',df2['keywords'].isnull().sum())
print('nan in tagline: ',df2['tagline'].isnull().sum())
# xử lý giá trị nan thành ký tự rỗng
df2['tagline'] = df2['tagline'].fillna('')
```

```python
import json
def get_info(x, num_feature=3):
    lst = []
    if x:
        for i in json.loads(x):
            lst.append(i['name'])
            if len(lst) == num_feature:
                break
    return ' '.join(lst)
```

```python
overviews = []
for i in range(len(df2['overview'])):
    overviews.append(' '.join([df2['title'][i], df2['overview'][i], get_info(df2['genres'][i]), 
                               get_info(df2['cast'][i]), get_info(df2['keywords'][i]), df2['tagline'][i]]))
```

```python
from nltk.corpus import stopwords
from nltk.stem.lancaster import LancasterStemmer
from pyvi import ViTokenizer

def Preprocessing(overview):
    overview = overview.lower()
    punctuation = '''!"#$%&\'()*+,.-/:;<=>?@[\\]^_`{|}~'''
    for punc in punctuation:
        if punc == "'":
            overview = overview.replace(punc, '')
        else: overview = overview.replace(punc, ' ')
    overview = ' '.join([word for word in overview.split() if word not in stopwords.words('english')])
    overview = ' '.join([LancasterStemmer().stem(word) for word in overview.split()])
    return overview
```

```python
overviews = [Preprocessing(overview) for overview in overviews]
```

```python
overviews[:5]
```

```python
from keras.preprocessing.text import Tokenizer
from keras.utils import pad_sequences

vocab_size = 18591 # số lượng từ trong tập từ vựng thống kê theo countvectorizer in sklearn
oov_tok = "<OOV>"
tokenizer = Tokenizer(num_words=vocab_size, oov_token=oov_tok, filters='')

tokenizer.fit_on_texts(overviews)
print("Number of Documents: ", tokenizer.document_count)
print("Number of Words: ", tokenizer.num_words)
```

```python
tokenizer.index_word
```

```python
sequences = tokenizer.texts_to_sequences(overviews)
```

```python
max_len = 0
sum = 0
for x in sequences:
  sum += len(x)
  if len(x) > max_len:
    max_len = len(x)
print('Câu có độ dài lớn nhất: ', max_len)
print('Độ dài trung bình câu: ', sum/len(sequences))
```

```python
sequence_length = 142
sequences_padded = pad_sequences(sequences, maxlen=sequence_length, padding='post', truncating='post')
```
